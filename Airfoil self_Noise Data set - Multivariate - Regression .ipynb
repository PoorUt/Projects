```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%
```

```python
column_names = ["Frequency(Hz)","Angle of Attack(Degree)","Chord length(meters)","Free-stream velocity(meters per second)",
                     "Suction side displacement thickness (meters)", "Scaled sound pressure level(decibels)"]
df=pd.read_csv('airfoil_self_noise.dat', delimiter = '\t', header=None)
df
```

Data Set Information:

The NASA data set comprises different size NACA 0012 airfoils at various wind tunnel speeds and angles of attack. The span of the airfoil and the observer position were the same in all of the experiments.


Attribute Information:

This problem has the following inputs:
1. Frequency, in Hertzs.
2. Angle of attack, in degrees.
3. Chord length, in meters.
4. Free-stream velocity, in meters per second.
5. Suction side displacement thickness, in meters.

The only output is:
6. Scaled sound pressure level, in decibels.

```python
#Lets define X and y parameters
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state = 42)
```

#Since there are so many data points with continous variable we can go with regression algorithms
# Options we have:
    #Linear regression
    #SVR
    #Ridge, Lasso, elastic net
    #Decision tree regression
    

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
import pickle
pickle.dump(regressor,open('model.pkl','wb'))
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
```

```python
r2
```

# Lets do ridge regression and check model accuracy

```python

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()
ridge
```

```python
parameters = {'alpha':[1e-15,1e-3,1e-2,1,5,10,20,30,40,50]}
```

```python
ridge_regressor = GridSearchCV(ridge,parameters,scoring = 'neg_mean_squared_error',cv=5, verbose=1)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
```

```python
print(ridge_regressor.best_score_)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor = GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)

lasso_regressor.fit(X_train,y_train)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

# Elastic Net

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
elastic_net = ElasticNet()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
elasticnet_regressor = GridSearchCV(elastic_net,parameters,scoring='neg_mean_squared_error',cv=5)

elasticnet_regressor.fit(X_train,y_train)
print(elasticnet_regressor.best_params_)
print(elasticnet_regressor.best_score_)
```
